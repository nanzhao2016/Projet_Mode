```python
def get_images_in_folder(folder_name):
    sub_folders = os.listdir(path) 
    data_folders = [os.path.join(folder_name, folder) for folder in sub_folders]
    folder_names = []
    image_number = []
    image_bytes = []
    for folder in data_folders:
        image_files = os.listdir(folder)
        folder_names.append(folder.split('store/')[1])
        image_number.append(len(image_files))
        size = 0
        for image in image_files:
           size = size + os.path.getsize(os.path.join(folder, image))
        image_bytes.append(size)
    return(folder_names, image_number, image_bytes)
```

```python
folder_names, image_number, image_bytes = get_images_in_folder(path)
```

```python
images_draw = pandas.DataFrame({'motif': folder_names, 'number':image_number, 'bytes':image_bytes})
images_draw.head()
```

```python
def rice_interval(data, column):
    len_interval = round(math.pow(len(data), 1/3)*2)
    return (round((max(data[column])-min(data[column]))/len_interval))
rice_interval(images_draw, 'number')
```
