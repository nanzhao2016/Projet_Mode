```python
def get_images_in_folder(folder_name, strsplit):
    sub_folders = os.listdir(folder_name) 
    data_folders = [os.path.join(folder_name, folder) for folder in sub_folders]
    folder_names = []
    image_number = []
    image_bytes = []
    for folder in data_folders:
        image_files = os.listdir(folder)
        folder_names.append(folder.split(strsplit)[1])
        image_number.append(len(image_files))
        size = 0
        for image in image_files:
           size = size + os.path.getsize(os.path.join(folder, image))
        image_bytes.append(size)
    return(folder_names, image_number, image_bytes)
```

```python
def rice_interval(data, column):
    len_interval = round(math.pow(len(data), 1/3)*2)
    return (round((max(data[column])-min(data[column]))/len_interval))
rice_interval(images_draw, 'number')
```

```python
import shutil
def mergeImageFolder(sourceRoot, destRoot):
    if not os.path.exists(destRoot):
        return False
    ok = True
    for path, dirs, files in os.walk(sourceRoot):
        print("path, dirs: ", path, dirs) 
        print(len(files), len(dirs))
        relPath = os.path.relpath(path, sourceRoot)
        print("relPath: ", relPath)
        destPath = os.path.join(destRoot, relPath)
        print("destPath: ", destPath)
        if not os.path.exists(destPath):
            os.makedirs(destPath)
        for file in files:
            destFile = os.path.join(destPath, file)
            if os.path.isfile(destFile):
                print ("Skipping existing file: " , os.path.join(relPath, file))
                ok = False
                continue
            srcFile = os.path.join(path, file)
            #os.rename(srcFile, destFile)
            shutil.copy2(srcFile, destFile)
    for path, dirs, files in os.walk(sourceRoot):
        print("path, dirs: ", path, dirs)
        print(len(files), len(dirs))
        """
        if len(files) == 0 and len(dirs) == 0:
            os.rmdir(path)
        """
    return (ok)
```

```python
def merge_image_folders(sourceRoot, destRoot):
    if not os.path.exists(destRoot):
        os.makedirs(destRoot)
    ok = True
    for path, dirs, files in os.walk(sourceRoot):
        for file in files:
            destFile = os.path.join(destRoot, file)
            if os.path.isfile(destFile):
                print ("Skipping existing file: " , os.path.join(path, file))
                ok = False
                continue
            srcFile = os.path.join(path, file)
            shutil.copy2(srcFile, destFile)
    return(ok)
```
